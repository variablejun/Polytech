```
from google.colab.patches import cv2_imshow
import numpy as np
import cv2
import matplotlib.pyplot as plt
from google.colab import drive
#라플라시안 검출기
#기존 검출기들은 1차 미분연산에 기반한것, 라플라시안 검출기는 2차미분연산을 이용하여 에지검출
#라플라시안은 2차미분 결과가 부호가 바뀌는 지점인 0에위치에서 검출
# 0이 교차하는 지점에서 에지검출
# 잡음에 예민
drive.mount('/content/gdrive') 
#영상 불러오기
#img = cv2.imread('/content/gdrive/My Drive/Image_Processing/Lena-Gaussian-noise1.jpg') 
# 노이즈가 있을때 성능이 좋지않다. 
img = cv2.imread('/content/gdrive/My Drive/Image_Processing/lena.jpg')

# color영상을 gray영상으로 만들기
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# 라플라시안 에지 컨벌루션 연산하기
laplacian = cv2.Laplacian(gray_img,-1,1)
# src ddepth ksize인데 1을 설정한 경우 3*3이다
output_img = laplacian/laplacian.max()

#그림을 화면에 출력
plt.figure(figsize=(20,20))# 영상의 크기를 키워주자
plt.subplot(1,2,1)
plt.title("Image1")
plt.imshow(gray_img, cmap='gray')
plt.axis("off")
plt.subplot(1,2,2)
plt.title("Laplacian Edge Detection")
plt.imshow(output_img, cmap='gray')
plt.axis("off")
plt.show()
```
