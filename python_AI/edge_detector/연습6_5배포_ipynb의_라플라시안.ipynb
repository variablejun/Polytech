```
from google.colab.patches import cv2_imshow
import numpy as np
import cv2
import matplotlib.pyplot as plt
from google.colab import drive
# 
#  1  ,   2  
# 2     0 
# 0   
#  
drive.mount('/content/gdrive') 
# 
#img = cv2.imread('/content/gdrive/My Drive/Image_Processing/Lena-Gaussian-noise1.jpg') 
#    . 
img = cv2.imread('/content/gdrive/My Drive/Image_Processing/lena.jpg')

# color gray 
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

#    
laplacian = cv2.Laplacian(gray_img,-1,1)
# src ddepth ksize 1   3*3
output_img = laplacian/laplacian.max()

#
plt.figure(figsize=(20,20))#
plt.subplot(1,2,1)
plt.title("Image1")
plt.imshow(gray_img, cmap='gray')
plt.axis("off")
plt.subplot(1,2,2)
plt.title("Laplacian Edge Detection")
plt.imshow(output_img, cmap='gray')
plt.axis("off")
plt.show()
```
