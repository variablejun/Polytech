```
from google.colab.patches import cv2_imshow
import numpy as np
import cv2
import matplotlib.pyplot as plt
from google.colab import drive
#캐니 에지 검출기
#1차 미분연산자인 로버트, 소벨, 프르윗, 검출기 단점 보완
# 과정 가우시안 커널을 이용해 영상잡음제거 -> 에지 검출기를 이용하여 검출
# -> 2개의 임계값을 이용한 에지 트래킹을 통해 강한 에지만을 검출

# 에지 트레킹
# 그래디언트값 t1설정하고 그것보다 큰 픽셀을 찾아 시작점으로 설정
# 시작점으로부터 양방향으로t2 값보다 작은 그레디언트값이 나올때 까지 트래킹하여 에지검출
drive.mount('/content/gdrive') 
#영상 불러오기
#img = cv2.imread('/content/gdrive/My Drive/Image_Processing/Lena-Gaussian-noise1.jpg')
img = cv2.imread('/content/gdrive/My Drive/Image_Processing/lena.jpg')

# color영상을 gray영상으로 만들기
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# 케니 에지 컨벌루션 연산하기
# T1: 100, T2: 250

output_img = cv2.Canny(gray_img,200,100) # src t1 t2
# 적절하게 높은 임계값(T1), 너무 작으면 원하지 않는 에지들이 포함되고 너무 크면 잘 검출하지 못함
# 낮은 임계값(T2) 너무 크면 서리 이어지지 않고 끊기게 되고 너무 작으면 원하지 않은 에지가 포함됨

#그림을 화면에 출력
plt.figure(figsize=(20,20))# 영상의 크기를 키워주자
plt.subplot(1,2,1)
plt.title("Image1")
plt.imshow(gray_img, cmap='gray')
plt.axis("off")
plt.subplot(1,2,2)
plt.title("Canny Edge Detection")
plt.imshow(output_img, cmap='gray')
plt.axis("off")
plt.show()
```
