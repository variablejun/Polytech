```
from google.colab.patches import cv2_imshow
import numpy as np
import cv2
import matplotlib.pyplot as plt
from google.colab import drive

drive.mount('/content/gdrive') 
#영상 불러오기
img = cv2.imread('/content/gdrive/My Drive/Image_Processing/airplane.bmp')

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

height, width = img.shape[0], img.shape[1]
#Flip
output1 = np.zeros((height, width,3), dtype=np.uint8)
output2 = np.zeros((height, width,3), dtype=np.uint8)

for h in range(height):
    for w in range(width):
        hp = height - h - 1       #상하 반전
        wp = width - w - 1        #좌우 반전
        output1[h,w] = img[hp, w] #상하 반전
        output2[h, w] = img[h, wp] #좌우 반전

#그림을 화면에 출력
plt.figure(figsize=(20,20))# 영상의 크기를 키워주자
plt.subplot(1,3,1)
plt.title("Original image")
plt.imshow(img)
plt.axis("off")
plt.subplot(1,3,2)
plt.title("Flipped vertical image")
plt.imshow(output1)
plt.axis("off")
plt.subplot(1,3,3)
plt.title("Flipped horizontal image")
plt.imshow(output2)
plt.axis("off")
```
