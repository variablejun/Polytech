```
from google.colab.patches import cv2_imshow
import numpy as np
import cv2
import matplotlib.pyplot as plt
from google.colab import drive
'''
대칭 변환
영상을 특정 축이나 점에 대해 반사시키는 것
'''
drive.mount('/content/gdrive') 
#영상 불러오기
img = cv2.imread('/content/gdrive/My Drive/Image_Processing/airplane.bmp')
# BGR영상을 RGB영상으로 변환
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
'''
입력 영상, 0 가로축기준으로, 1 세로축 기준으로 ,-1은 두축 을 중심으로 대칭시킴
'''
flipVertical = cv2.flip(img, 0) #가로축 기준
flipHorizontal = cv2.flip(img, 1)  #세로축 기준
flipBoth = cv2.flip(img, -1)  #두축 중심으로 대칭
 
#그림을 화면에 출력
plt.figure(figsize=(20,20))# 영상의 크기를 키워주자
plt.subplot(1,4,1)
plt.title("Original image")
plt.imshow(img)
plt.axis("off")
plt.subplot(1,4,2)
plt.title("Flipped vertical image")
plt.imshow(flipVertical)
plt.axis("off")
plt.subplot(1,4,3)
plt.title("Flipped horizontal image")
plt.imshow(flipHorizontal)
plt.axis("off")
plt.subplot(1,4,4)
plt.title("Flipped both image")
plt.imshow(flipBoth)
plt.axis("off")
plt.show()
```
